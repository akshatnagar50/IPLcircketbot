```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv("../input/powerplay/Copy of IPL_2020.csv")
```

```python
data
```

```python
encode = {'batting_team':{'RCB':1,'CSK':2,'DC':3,'MI':4,'SRH':5,'KKR':6,'PBKS':7,'RR':8,'PW':9,'KT':10,'GL':11,'DN':12},
          'bowling_team':{'RCB':1,'CSK':2,'DC':3,'MI':4,'SRH':5,'KKR':6,'PBKS':7,'RR':8,'PW':9,'KT':10,'GL':11,'DN':12}}
data.replace(encode,inplace=True)
```

```python
data.head()
```

```python
#converting categorical data to numerical data
from sklearn.preprocessing import LabelEncoder
var_mod=['venue','Pitch Type']
le=LabelEncoder()
for i in var_mod:
   data[i]=le.fit_transform(data[i])
```

```python
data
```

```python
#Machine Learning Model
x=data.iloc[:,[0,1,2,3,5,6]].values
y=data.iloc[:,[4]].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
#ML Model
from sklearn.linear_model import LinearRegression
regr=LinearRegression()
regr.fit(x_train,y_train)
```

```python
#custom input
import numpy as np
array=list(map(int, input().split()[:6]))
new_prediction=regr.predict(sc.transform([array]))
a=np.round(new_prediction)
print("Prediction Score:",a)
```
